Question # 1

```python
input1 = int(input("Enter first value: "))
input2 = int(input("Enter second value: "))
operand = input("Enter operand (+,-,*,/,^): ")

if(operand == "+"):
    print(input1 + input2)
elif(operand == "-"):
    print(input1 - input2)
elif(operand == "/"):
    print(input1 / input2)
elif(operand == "x"):
    print(input1 * input2)
elif(operand == "^"):
    print(input1 ** input2)
else:
    print("Plz enter correct operand")
```

Question # 2

```python
my_list = ["apple", "banana", 10, "cat"]

for item in my_list:
    if type(item) == int or type(item) == float:
        print("Numeric value found:", item)
        break
else:
    print("No numeric value found")
```

Question # 4

```python
my_dict = {"a": 10, "b": 20, "c": "hello", "d": 5}
total = 0
for value in my_dict.values():
    if type(value) == int or type(value) == float:
        total += value

print("Total sum:", total)
```

Question # 5

```python
my_list = [1, 2, 3, 2, 4, 5, 1]
duplicates = []

for item in my_list:
    if my_list.count(item) > 1 and item not in duplicates:
        duplicates.append(item)

print("Duplicate values:", duplicates)
```
