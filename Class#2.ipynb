 # Marksheet App

```python
Ds = int(input("Enter Datascience Marks:"))
Ml = int(input("Enter Machine learning Marks:"))
Os = int(input("Enter Operating System Marks:"))
Ata = int(input("Enter Automata Marks:"))

obtain = Ds + Ml + Os + Ata
print("Your Obtained Marks is : ",obtain)

perc = obtain / 400 * 100
print("Your Percentage is : ", perc)

if perc >=80 and perc <=100:
    print("A-One Grade")
elif perc >= 70 and perc < 80:
    print("A Grade")
elif perc >= 60 and perc < 70:
    print("B Grade")
elif perc >= 50 and perc < 60:
    print("C Grade")
elif perc >= 40 and perc < 50:
    print("D Grade")
elif perc >= 33 and perc < 40:
    print("E Grade")
else:
    print("Fail")
```

# Calculator App

```python
input1 = int(input("Enter first value: "))
input2 = int(input("Enter second value: "))
operand = input("Enter operand (+,-,*,/): ")

if(operand == "+"):
    print(input1 + input2)
elif(operand == "-"):
    print(input1 - input2)
elif(operand == "/"):
    print(input1 / input2)
elif(operand == "x"):
    print(input1 * input2)
else:
    print("Plz enter correct operand")
```
